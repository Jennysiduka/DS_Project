```python
import pandas as pd
pd.set_option('display.precision',7)
star_wars=pd.read_csv('star_wars.csv',encoding='ISO-8859-1')
```

```python
star_wars.columns
```

```python
star_wars.head()
```

```python
star_wars=star_wars[pd.notnull(star_wars['RespondentID'])]
```

```python
star_wars['RespondentID']= star_wars['RespondentID'].apply(lambda x : int(x))
```

```python
yes_no = {'Yes': True,'No': False}
for col in ['Have you seen any of the 6 films in the Star Wars franchise?','Do you consider yourself to be a fan of the Star Wars film franchise?']:
    star_wars[col]= star_wars[col].map(yes_no)
    
star_wars.head()
```

```python
#Investigate the value of the columns
for col in star_wars.columns[3:9]:
    print(star_wars[col].value_counts())
```

```python
import numpy as np
transfer_names = {'Star Wars: Episode I  The Phantom Menace':True,'Star Wars: Episode II  Attack of the Clones': True,'Star Wars: Episode III  Revenge of the Sith':True,'Star Wars: Episode IV  A New Hope':True,'Star Wars: Episode V The Empire Strikes Back':True,'Star Wars: Episode VI Return of the Jedi':True,np.NaN:False}
for col in star_wars.columns[3:9]:
    star_wars[col]=star_wars[col].map(transfer_names)
    
star_wars=star_wars.rename(columns= {'Which of the following Star Wars films have you seen? Please select all that apply.':'seen_1','Unnamed: 4': 'seen_2','Unnamed: 5':'seen_3','Unnamed: 6':'seen_4','Unnamed: 7':'seen_5','Unnamed: 8':'seen_6'}
)
star_wars.head()
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
cols={}
for i in range(6):
    cols[star_wars.columns[i+9]] = 'rank_{}'.format(i+1)

star_wars=star_wars.rename(columns=cols)
```

```python
#convert columns names
star_wars.columns[9:15]
```

```python
%matplotlib inline
star_wars[star_wars.columns[9:15]].mean()
star_wars[star_wars.columns[9:15]].mean().plot.bar()
```

The fifth Star Wars: Empire Strikes Back has the lowest mean score, which means that the respondents like this movie the most.

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
star_wars[star_wars.columns[3:9]].sum().plot.bar()
```

We can see that the 5th star_wars has the most number of audience. The number also matches the ranking score of each movie. It seems the more favorite the movie is, the more audience the movie has.

```python
import matplotlib.pyplot as plt
males = star_wars[star_wars['Gender']=='Male']
females = star_wars[star_wars['Gender']=='Female']
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
ax1.bar(range(1,7), males[star_wars.columns[9:15]].mean())
ax1.set_title('Male / Star Wars Ranking Score')
ax1.set_xticks(range(1,7))
ax2.bar(range(1,7), males[star_wars.columns[3:9]].sum())
ax2.set_title('Male / Total Audience')
ax2.set_xticks(range(1,7))
ax3.bar(range(1,7), females[star_wars.columns[9:15]].mean())
ax3.set_title('Female / Star Wars Ranking Score')
ax3.set_xticks(range(1,7))
ax4.bar(range(1,7), females[star_wars.columns[3:9]].sum())
ax4.set_title('Female / Total Audience')
ax4.set_xticks(range(1,7))
plt.show()
```

Regarding to the movies' preference, males like the last 3 episodes more than the 1st 3 episodes,while females like the 1st and the last 2 episodes the most.
1st and the last 2 episodes have the most number of male and female audience. Male like star_wars more than female.
